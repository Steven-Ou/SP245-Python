```python
import matplotlib.pyplot as plt
import numpy as np

# Define vector
v1 = np.array([2, 3])

# Create plot
fig, ax = plt.subplots(figsize=(6, 6))

# Draw x and y axes
ax.axhline(0, color='black', linewidth=1)  # X-axis (horizontal)
ax.axvline(0, color='black', linewidth=1)  # Y-axis (vertical)

# Plot the vector
ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r')

# Set limits and grid
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_xticks(range(-5, 6))
ax.set_yticks(range(-5, 6))
ax.grid(True, linestyle='--', linewidth=0.5)

# Labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('2D Vector Plot')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define vectors
vectors = {
    r'$\mathbf{v_1}$': np.array([1, 5, 0]),
    r'$\mathbf{v_2}$': np.array([5, 1, 0]),
    r'$\mathbf{v_3}$': np.array([0, 0, 4]),
    r'$\mathbf{v_4}$': np.array([5, 5, 5])
}

# Create a 3D figure
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# Draw x, y, z axes
ax.quiver(0, 0, 0, 1, 0, 0, color='black', arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, 0, 1, 0, color='black', arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, 0, 0, 1, color='black', arrow_length_ratio=0.1)

# Plot vectors
for label, vec in vectors.items():
    ax.quiver(0, 0, 0, vec[0], vec[1], vec[2], color=np.random.rand(3,))
    ax.text(vec[0], vec[1], vec[2], label, fontsize=12)

# Set axis limits
ax.set_xlim([0, 6])
ax.set_ylim([0, 6])
ax.set_zlim([0, 6])

# Labels
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Vector Plot')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the vectors
v1 = np.array([1, -4])
v2 = np.array([-2, 1])
v3 = np.array([-3, -4])
v4 = np.array([3, 4])

# Create the plot
fig, ax = plt.subplots()
# Draw x and y axes
ax.axhline(0, color='black', linewidth=1)  # X-axis (horizontal)
ax.axvline(0, color='black', linewidth=1)  # Y-axis (vertical)

# Plot the vectors
ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label='v1')
ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='g', label='v2')
ax.quiver(0, 0, v3[0], v3[1], angles='xy', scale_units='xy', scale=1, color='b', label='v3')
ax.quiver(0, 0, v4[0], v4[1], angles='xy', scale_units='xy', scale=1, color='purple', label='v4')

# Set the axis limits
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)

# Add grid, labels, and title
ax.grid(True)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Vector Plot')

# Add legend for vector labels
ax.legend()

# Set aspect ratio to be equal
ax.set_aspect('equal')

# Show the plot
plt.show()
```
