```python
import matplotlib.pyplot as plt
import numpy as np

# Define vector
v1 = np.array([2, 3])

# Create plot
fig, ax = plt.subplots(figsize=(6, 6))

# Draw x and y axes
ax.axhline(0, color='black', linewidth=1)  # X-axis (horizontal)
ax.axvline(0, color='black', linewidth=1)  # Y-axis (vertical)

# Plot the vector
ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r')

# Set limits and grid
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_xticks(range(-5, 6))
ax.set_yticks(range(-5, 6))
ax.grid(True, linestyle='--', linewidth=0.5)

# Labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('2D Vector Plot')

plt.show()
```
